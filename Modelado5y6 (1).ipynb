```python
import numpy as np
import sympy as sp
import math  as mt
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from pylab import *
```

```python
xmax = 5
xmin = -xmax

ymax = 5
ymin = -ymax

Nx=10
Ny=10

x=np.linspace(xmin, xmax, Nx)
y=np.linspace(ymin, ymax, Nx)

X, Y= meshgrid(x, y)

Bx= -Y
By= -X

figure(figsize=[5,5])

QP = quiver(X, Y, Bx, By,linewidth=0.5, color= 'r' )
quiverkey(QP, 0.85, 1.05, 1, '1 mt', labelpos='N')
dx = (xmax-xmin)/(Nx-1)
dy = (ymax-ymin)/(Ny-1)
axis([xmin-dx, xmax+dx, ymin-dy, ymax+dy])

def fun(t,xy):
    X,Y=xy
    return [-Y,-X]

x0 = np.arange(-5,5,1)
y0 = np.arange(-5,5,1)
ti=0; tf=1; dt=20
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for i in range(len(x0)):
    for j in range(len(y0)):
        sol=solve_ivp(fun,tspam,[x0[i],y0[j]],t_eval=t)
        plt.plot(sol.y[1], sol.y[0])
plt.axis([-5, 5, -5, 5])
plt.grid()
plt.plot(0,0, marker="o", color="green")

title('Campo vetcorial')
xlabel('x')
ylabel('y')
show()
```
